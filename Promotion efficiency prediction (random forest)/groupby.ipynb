# groupby 연습하기

```python
import pandas as pd
```

```python
data = pd.read_csv('transaction.csv')
```

```python
data = data.head(1000)
```

```python
data.groupby('id').sum()
```

```python
data.groupby('id').mean()
```

```python
data.groupby('id').max()
```

```python
data.groupby('id').count()
```

```python
data.groupby('id').describe()
```

```python
data.set_index('id', inplace=True)
```
