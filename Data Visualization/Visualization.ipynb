# 데이터 시각화 (Visualization) - Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 10) # 시작,끝,등분 갯수 
```

```python
y = x ** 2
```

```python
plt.plot(x,y,'r') #'색상코드' /r => red , g => green 
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig = plt.figure() #빈 캠퍼스 생성

axes = fig.add_axes([0.1,0.1,0.7,0.7])

axes.plot(x,y,'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
plt.plot(x,y,'g') #'색상코드' /r => red , g => green 
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
plt.plot(y,x,'b') #'색상코드' /r => red , g => green 
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.2,0.6,0.3,0.3])

axes1.plot(x,y,'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')

axes2.plot(y,x,'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig, axes = plt.subplots() #여러 plot 그릴때 사용함
axes.plot(x,y,'g')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Plot Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')

axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Title_1')

axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Title_2')
```

```python
fig.savefig('my_name.png')  #png으로 저장됨
```

# 데이터 시각화 (Visualization) - Distribition Plot

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.distplot(data['total_bill'])
```

```python
sns.distplot(data['total_bill'], kde =False)   #Distribition Plot으로 변경
```

```python
sns.distplot(data['total_bill'], kde =False,bins =10)   #Distribition Plot으로 변경
```

```python
sns.jointplot(x = 'total_bill', y='tip', data= data)
```

```python
sns.jointplot(x = 'total_bill', y='tip', data= data, kind='hex')
```

```python
sns.jointplot(x = 'total_bill', y='tip', data= data, kind='reg')
```

- jointplot()은 두 변수를 각각 relational plot과 distribution plot로 나타내서 한번에 그려주는 함수입니다. 기본값으로는 scatterplot()과 histplot()을 한번에 나타냅니다. 그리고 jointplot()은 data가 키워드타입이기 때문에 'data=penguins'로 키워드 대입으로 매개변수를 입력해주어야 합니다.


```python
sns.pairplot(data)
```

- pairplot : 데이터에 들어 있는 각 컬럼(열)들의 모든 상관 관계를 출력

3차원 이상의 데이터라면 pairplot 함수를 사용해 분포도를 그립니다.
pairplot은 그리도(grid) 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 그립니다

```python
sns.pairplot(data, hue = 'smoker') #smoker가 yes,no에 따라 색깔 다름 
```

```python
sns.rugplot(data['total_bill']) #바코드 plot 
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data['time'].unique()
```

```python
g = sns.FacetGrid(data, col='time',row ='smoker')
g = g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(data, col='time',row ='smoker')
g = g.map(plt.scatter, 'total_bill', 'tip')
```

```python
g = sns.FacetGrid(data, col='time',row ='smoker', hue= 'sex') #성별에 따라 분포됨
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data =data, hue='sex')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data =data, hue='sex',palette='Set2')
```

```python
data.head()
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data =data, hue='sex',palette='Set2', col='time')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data =data, hue='sex',palette='Set2', col='time', row= 'smoker')
```
